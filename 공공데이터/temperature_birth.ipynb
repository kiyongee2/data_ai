```python
import csv
import matplotlib.pyplot as plt

with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)
    next(data)
    temp_high = []   # 최고 기온 
    temp_low = []    # 최저 기온

    for row in data:
        if row[-1] != "" and row[-2] != "":  # 최고 기온과 최저 기온이 존재한다면
            if int(row[0].split('-')[0]) >= 1980:  # 1980년 이후 데이터라면
                if row[0].split('-')[1] == '07' and row[0].split('-')[2] == '20':
                    temp_high.append(float(row[-1]))
                    temp_low.append(float(row[-2]))

plt.rc('font', family='Malgun Gothic')      #한글 글꼴 추가
plt.rcParams['axes.unicode_minus'] = False  #마이너스 기호 깨짐 방지
plt.title("내 생일의 기온 변화 그래프")
plt.plot(temp_high, 'red', label='high')
plt.plot(temp_low, 'blue', label='low')
plt.xlabel("연도")
plt.ylabel("기온")
plt.legend()
plt.show()
```
