```python
import urllib.request

url = "http://www.kopis.or.kr/openApi/restful/prfplc?service=fd1d23255076491484dd7d805ccc7dce&cpage=1&rows=5&shprfnmfct=예술의전당"

# 요청
request = urllib.request.Request(url)  #url주소를 요청(호출)
# 응답
response = urllib.request.urlopen(request)   #requests.get(url) 유사함
print(response.getcode())  # 상태 코드 가져옴 200, 404
print(type(response))
result = response.read().decode('utf-8')
print(result)
```
