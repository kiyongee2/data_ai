```python
import pandas as pd

df = pd.read_csv("./datas/auto-mpg0.csv", header=None)
df
# 열 이름 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',  
        'acceleration', 'model_year', 'origin', 'car_name']
df.head()
```

```python
df.shape
df.dtypes
df.info()
```

```python
df.describe()
df.describe(include='all')
df.describe(include='number')
df.describe(include='object')
```

```python
df.count()
type(df.count())

# 제조국 열이 가지는 고유값의 개수
df['origin'].value_counts()
df['origin'].value_counts(normalize=True)
# 백분율
df['origin'].value_counts(normalize=True) * 100
```

```python
# 평균값
df.mean(numeric_only=True)
df['mpg'].mean()
df[['mpg', 'weight']].mean()

# 중앙값
df.median(numeric_only=True)
df['mpg'].median()

# 최대값
```

```python
# 상관계수(Correlation Coefficient)
df.corr(numeric_only=True)
df[['mpg', 'weight']].corr()
```

```python
# 판다스 내장 그래프
df = pd.read_excel("./datas/남북한발전전력량.xlsx")
df

# 남북한 발전량 합게 데이터만 추출
df_ns = df.iloc[[0, 5], 3:]
df_ns
df_ns.index = ["South", "North"]
# df_ns.info()

df_ns = df_ns.astype(int)
df_ns.info()
df_ns.head()
```

```python
# 그래프 그리기
df_ns.plot()

# 행열 바꾸기
tdf_ns = df_ns.T
tdf_ns.head()

tdf_ns.plot()
```

```python
# 세로 막대 그래프
tdf_ns.plot(kind='bar')
tdf_ns.plot(kind='bar', stacked=True)

# 가로 막대 그래프
tdf_ns.plot(kind='barh')
```

```python
tdf_ns.plot(kind='hist')

tdf_ns.plot(kind='hist',
           alpha=0.5,
           grid=True,
           color='skyblue',
           edgecolor='black')
```

```python
tdf_ns.hist(figsize=(6, 2))

tdf_ns.hist(figsize=(6, 2), bins=20, grid=False)
```
