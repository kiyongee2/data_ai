```python
import folium
import pandas as pd

data = pd.read_csv("dataset/택시승차대 현황.csv", encoding='cp949')
data.head()

taxi = data.loc[:, ["위치명", "위도", "경도"]]
m = folium.Map(location=[37.547647896577175, 126.94245451708481], zoom_start=13, width=500, height=500)

# taxi.info()

for index, row in taxi.iterrows():
    folium.Marker([row["위도"], row["경도"]], 
                  tooltip=row["위치명"], 
                  icon=folium.Icon(color='red', icon='fa-solid fa-taxi', prefix='fa-solid')
                 ).add_to(m)
    folium.Circle([row["위도"], row["경도"]],
                  color='red',
                  radius=50
                 ).add_to(m)
#     print(index)
#     print(row)

m
```
