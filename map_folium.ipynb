```python
import folium

# 구글 지도에서 위도, 경도 가져옴
map = folium.Map(location=[37.54322, 126.951576], zoom_start=15)
# html 파일로 저장
map.save("./datas/map.html")
map
```

```python
# 마커 표시
folium.Marker(
  location=[37.54322, 126.951576],
  popup="공덕역", icon=folium.Icon(color='red', icon='star')
).add_to(map)
map
```

```python
# 주변 영역 표시
folium.CircleMarker(
    location=[37.54322, 126.951576],
    radius = 100,
    color = "#fff",
    popup="공덕역", icon=folium.Icon(color='red', icon='star')
).add_to(map)
map
```
