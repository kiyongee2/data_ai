```python
import matplotlib.pyplot as plt

month=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales_2019=[100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020=[90,110,130,120,140,160,170,160,150,180,200,190]

plt.title("Monthly Sales Comparison (2019-2020)") # 

#     
plt.plot(month, sales_2019, label="2019")
plt.plot(month, sales_2020, label="2020")

plt.legend() #  

#   
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories=["Category1","Category2","Category3","Category4","Category5"]
data=[20,35,15,27,45]

# ,   
plt.title("Bar Chart")
plt.bar(categories,data)

# 
plt.xlabel("Categories")
plt.ylabel("Values")

# 
plt.grid(axis="both")

# y  50 
plt.yticks([0,10,20,30,40,50],[0,10,20,30,40,50])

# x  
plt.xticks(rotation=45)

plt.show()
```
