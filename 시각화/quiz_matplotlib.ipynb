```python
import matplotlib.pyplot as plt

month=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales_2019=[100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020=[90,110,130,120,140,160,170,160,150,180,200,190]

plt.title("Monthly Sales Comparison (2019-2020)") # 제목설정

# 기본차트 그리기 및 레이블을 설정
plt.plot(month, sales_2019, label="2019")
plt.plot(month, sales_2020, label="2020")

plt.legend() # 레이블 보여주기

# 축 이름 설정
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories=["Category1","Category2","Category3","Category4","Category5"]
data=[20,35,15,27,45]

# 제목짓기, 기본 차트 그리기
plt.title("Bar Chart")
plt.bar(categories,data)

# 축이름
plt.xlabel("Categories")
plt.ylabel("Values")

# 격자
plt.grid(axis="both")

# y 축 50까지 보이게
plt.yticks([0,10,20,30,40,50],[0,10,20,30,40,50])

# x축 글씨 기울이기
plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt

labels=["Apple","Banana","Mango","Blueberry"]
sizes=[15,30,45,10]
explode=(0,0.1,0,0) # 튜플 아니고 리스트로 줘도 상관없음

# 차트그리기
"""
색상은 기본
"""
# startangle 변경 (파이차트 기울이기)
# counterclock=True (기본 반시계), 시계방향으로 돌리고 싶으면 counterclock False로
# 퍼센트와 소수점 아래 하나 보여주기
# 그림자 보여주기
plt.pie(sizes, labels=labels, explode=explode, autopct="%.1f%%", shadow=True,  startangle=100)
plt.show()
```
