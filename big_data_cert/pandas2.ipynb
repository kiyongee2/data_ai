```python
# 결측치 처리
df = pd.read_csv("cafe2.csv")
# df.info()
# print(df.isnull())
print(df.isnull().sum())

# 결측치 채우기
df["원산지"] = df['원산지'].fillna("코스타리카")
# df = df['원산지'].fillna("코스타리카")
# print(df)

# replace()
df = df.replace('아메리카노', '허브차')
# print(df)

update = {"허브차": "아메리카노", "녹차": "녹차라떼"}
df = df.replace(update)
print(df)

# loc / iloc
df.loc[6, "원산지"] = "대한민국"

df.loc[1:2, "이벤트가"] = 1000
print(df)

df.to_csv("cafe3.csv", index=False)
```

```python
# 문자열
import pandas as pd

df = pd.DataFrame({
    'A': ['데이터 분석', '기본 학습서', '퇴근 후 열공'],
    'B': [10, 20, 30],
    'C': ['ab cd', 'AB CD', 'ab cd ']
})

# str 속성 사용
df['A'] = df['A'].str.replace("분석", "시각화")
# 숫자인 경우
df['B'] = df['B'].replace(20, 200)
# print(df)

# print(df['A'].str.split())
print(df['A'].str.split()[1])
df['D'] = df['A'].str.split().str[0]
print(df)

df['A'].str.contains('학습서')
df["학습서포함유무"] = df['A'].str.contains('학습서')

df["문자길이"] = df['A'].str.len()

df["C"] = df["C"].str.lower()
df

df['C'] == "ab cd"

# 공백 제거
# df['C'] = df['C'].str.strip()

df['C'] = df['C'].str.replace(" ", "")
print(df['C'])
print(df['C'].str[1:3])

# C열의 1~2행 출력
print(df['C'][1:3])
```

```python
# 내장 함수
import pandas as pd

df = pd.read_csv("cafe3.csv")
print(df)

print(len(df))
print(df.shape[0])

# 조건에 맞는 합계(개수)
cond = df["가격"] > 5000
print(sum(cond))
print(len(df[cond]))

# 열별 합계
df.sum(numeric_only=True)

# 행별 합계 - 행과 열을 바꿈
df = df.T # T - transpose
df.sum(numeric_only=True, axis=1)
df
```

```python
import pandas as pd

df = pd.read_csv("cafe3.csv")
# print(df)

print("최대값: ", df['가격'].max())
print("최소값: ", df['가격'].min())
print("평균값: ", df['가격'].mean())
print("중앙값: ", df['가격'].median())
print("합계: ", df['가격'].sum())
print("표준편차: ", df['가격'].std())
print("분산: ", df['가격'].var())
```

```python
# 사분위 수
print(df)
print(df["가격"].quantile(.25))
print(df["가격"].quantile(.75))

# 1사분위수보다 작은 가격의 데이터
print(df[df["가격"].quantile(.25) > df["가격"]])

# 최빈값
print(df["원산지"].mode()[0])

print(df["가격"].max())
print(max(df["가격"]))

max_idx = df["가격"].idxmax()
print(max_idx)

print(df.loc[max_idx]["메뉴"])
print(df.loc[max_idx, "메뉴"])
print(df.iloc[max_idx, 0])

print(df.nlargest(3, "가격"))
print(df.nsmallest(3, "가격"))
```
