```python
# 데이터 시각화
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')  #한글 글꼴일때 명시
plt.title('기본 그래프 그리기 ')
plt.plot([10, 20, 30, 40])  #리스트 1개 y축값(결과값)
# plt.plot([1, 2, 3, 4], [10, 40, 20, 30]) #리스트 2개, x축값, y축값
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')  #한글 글꼴일때 명시
plt.title('범례 및 색상 표시하기') #제목
plt.plot([10, 20, 30, 40], label='asc', color='r') # 범례(label), red = r
plt.plot([40, 30, 20, 10], label='desc', color='g')
plt.legend() # 범례 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')  #한글 글꼴일때 명시
plt.title('x축 제목, y축 제목 표시') #제목
plt.plot([40, 30, 20, 10])
plt.xlabel("인구")
plt.ylabel("식량")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40], 'b.', label='circle') # 범례(label), red = r
plt.plot([40, 30, 20, 10], 'r^', label='triangle')
plt.legend() # 범례 표시
plt.show()
```

```python
# 막대 그래프
import matplotlib.pyplot as plt

print(range(6))
print(list(range(6)))

# bar(막대를 표시할 위치, 막대의 높이)
# plt.bar([1, 2, 3, 4], [6, 3, 2, 4])
plt.bar(range(6), [1, 2, 3, 5, 6, 7])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')  #한글 글꼴일때 명시
plt.title('히스토그램') #제목
plt.hist([1, 1, 2, 3, 4, 5, 5, 5, 5, 7])
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

dice = []
for i in range(10):
    dice.append(random.randint(1, 6))

print(dice)

plt.rc('font', family='Malgun Gothic')  #한글 글꼴일때 명시
plt.title('주사위 던지기') #제목
plt.hist(dice, bins=6) # bins는 가로축의 구간 개수
plt.show()
```
