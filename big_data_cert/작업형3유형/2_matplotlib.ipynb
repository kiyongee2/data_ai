```python
import matplotlib.pyplot as plt

plt.title("Title", pad=20, loc="center") # pad-여백

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]
plt.plot(x, y)
plt.plot([1, 2, 3, 4], [3, 5, 9, 7])

axis = {'both', 'x', 'y'}

# plt.grid()
plt.grid(True, axis='y')  # 격자형
plt.show()
```

```python
# 막대 그래프
import matplotlib.pyplot as plt

print(range(6))
print(list(range(6)))

# bar(막대를 표시할 위치, 막대의 높이)
plt.bar(range(6), [1, 2, 3, 5, 6, 7])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# x = np.arange(3)
# print(x)

x = range(3)

plt.bar(x, [100, 400, 900], color=['r', 'g', 'b'], width=0.6)

plt.xticks(x, ['2018', '2019', '2020'])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(3)

plt.barh(y, [100, 400, 900], color=['r', 'g', 'b'], height=0.6)

plt.xticks(y, ['2018', '2019', '2020'])

plt.show()
```

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.1, 0, 0.1]
colors = ['red', 'yellow', 'green', 'purple']
wegeprops = {'width': 0.8, 'edgecolor': 'b', 'linewidth': 0.5}

plt.pie(ratio, labels=labels, autopct='%.1f%%', 
        explode=explode, colors=colors, wedgeprops=wegeprops)

plt.show()
```
