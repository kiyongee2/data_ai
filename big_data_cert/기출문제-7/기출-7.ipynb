```python
import pandas as pd

df = pd.read_csv("student_assessment.csv")
# print(df.head())

# 1번 - 1
# df.info()
# print(df.isnull().sum())

# 결측 행 제거
df = df.dropna()
# print(df.isnull().sum())

# 학생이 가장 많이 수강한 과목
print(df['id_assessment'].value_counts())
max_id = df['id_assessment'].value_counts().idxmax()
print(max_id)

# print(df[df['id_assessment'] == 12]) #33개
result = (df['id_assessment'] == max_id)
df= df[result]

# 성적의 최고 점수
# print(df['score'].max())

# 수강생이 가장 많은 과목의 최고 점수
max_score = df[result]['score'].max()
print(max_score)

# 최고점수 표준화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['score'] = scaler.fit_transform(df[['score']])
print(round(df['score'].max(), 3))
```

```python
# 1번 - 2
import pandas as pd

df = pd.read_csv("stock_market.csv")
df.info()
print(df.head())

# close와 상관 계수 - 절대값
df_corr = df.corr()['close'].abs()
print(df_corr)

# 상관 관계가 높은 변수
col = df_corr.loc['DE1':'DE77'].idxmax()
print(col)

# 찾은 변수의 평균값
mean_val = df[col].mean()
print(round(mean_val, 4))
```

```python
# 1번 - 3
import pandas as pd

df = pd.read_csv("air_quality.csv")
print(df)

# IQR 계산
Q1 = df['CO2'].quantile(0.25)
Q3 = df['CO2'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)

# 상한 및 하한 계산
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# 이상치 식별
result = (df['CO2'] < lower) | (df['CO2'] > upper)
print(df[result])
outliers = df[result]
print(len(outliers))
```

### 2유형

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

train = pd.read_csv("mart_train.csv")
test = pd.read_csv("mart_test.csv")

# train.info()
# train.head()
# print(train.shape, test.shape)

# 원-핫 인코딩
train = pd.get_dummies(train)
test = pd.get_dummies(test)

# train.info()

# 변수 선택
X = train.drop("total", axis=1)
y = train["total"]
print(train.shape, test.shape)

# 학습용 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X,
        y, test_size=0.2, random_state=303)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)

# 모델 학습 및 평가 
model = LinearRegression()
# model = RandomForestRegressor(random_state=303)

model.fit(X_train, y_train)
pred = model.predict(X_test)
# pred

# 성능 평가 - 지표 : RMSE(평균 제곱근 오차)
def rmse(y_true, y_pred):
    return mean_squared_error(y_true, y_pred) ** 0.5

# rmse = rmse(y_test, pred)

# numpy 사용
mse = mean_squared_error(y_test, pred)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

# 평가용 데이터 예측 및 결과 파일
pred2 = model.predict(test)
# print(pred2)

result = pd.DataFrame({"pred": pred2})
result.to_csv("result.csv", index=False)

pd.read_csv("result.csv")
```

#### 3유형

```python
# 1 - 1번
import pandas as pd
from statsmodels.formula.api import logit
import numpy as np

df = pd.read_csv("clam.csv")
# df.info()
# df.head()

# 데이터셋 분할
train = df.iloc[:210]
test = df.iloc[210:]
print(train.shape, test.shape)

# 로지스틱 회귀 모델 생성 및 훈련 데이터로 학습
model = logit("gender ~ weight", train).fit()
print(model.summary())

print(model.params)
print(model.pvalues)
w_coef = model.params['weight']
# print(w_coef)

# 오즈비 계산
odds_ratio = np.exp(w_coef)
print(round(odds_ratio, 4))

# 1 - 2번

# 모델 학습
formula = "gender ~ age + length + diameter + height + weight"
model2 = logit(formula, train).fit()

# 잔치 이탈도
result = -2 * model2.llf
print(round(result, 2))

from sklearn.metrics import accuracy_score
# 1 -3번
# 1) test데이터를 사용해 예측 (0.5 미만: 0, 0.5 이상 1)
model = logit("gender ~ weight", data=train).fit()
target = test.pop('gender')
pred = model.predict(test)
# pred = (pred > 0.5).astype(int)
pred = pred.astype(int)

# 2) 실제 값과 예측값을 사용하여 정확도 계산
accuracy = accuracy_score(target, pred)
print(accuracy)

# 3) 오류율 계산
error_rate = 1 - accuracy
print(round(error_rate, 3))
```

```python
# 3 - 2번
import pandas as pd
from statsmodels.formula.api import ols

df = pd.read_csv("system_cpu.csv")
# df.info()
# df.head()

# ERP와 상관 관계가 높은 값
corr = df.corr()
max_val = corr['ERP'].max()
print(round(max_val, 3))

# CPU가 100미만인 데이터로 선형 회귀 모델 구축
cond = df['CPU'] < 100
new_df = df[cond]

formula = "ERP ~ Feature1 + Feature2 + Feature3 + CPU"
model = ols(formula, new_df).fit()
print(model.summary())

# 결정 계수
print("R2:", round(model.rsquared, 3))

# p-value가 가장 높은 값
max_val = model.pvalues.max()
print(round(max_val, 3))
```
