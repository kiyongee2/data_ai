```python
# 조건문
# 1. if구문
age = 14

if age >= 15:
    print("관람가 입니다.")
print("나이는 ", age, "세입니다.")

# 2. if ~ else 구문
if age >= 15:
    print("관람가 입니다.")
else:
    print("관람불가 입니다.")
print("나이는 ", age, "세입니다.")
```

```python
# 3. if ~ elif ~ else 구문
# 점수 = 87

점수 = input("점수 입력: ")
점수 = int(점수) #문자형을 정수형으로 변환해야 함. 
print(type(점수))

if 점수 >= 90:
    print('A 학점')
elif 점수 >= 80:
    print('B 학점')
elif 점수 >= 70:
    print('C 학점')
else:
    print('F 학점')
print("Score : ", 점수, "점")
```

```python
# 반복문
# while문
# 1부터 10까지 출력하기 
n = 1

while n <= 10:
    print(n)
    n = n + 1
    
# 1부터 10까지 더하기
n = 1
total = 0
while n <= 10:
    total = total + n
    print("n =", n, "total =", total)
    n = n + 1
    
print("합계는", total, "입니다")
```

```python
# 반복 조건문(while ~ if ~ break문)

n = 1

while True:
    if n > 10:
        break
    print(n)
    n = n + 1

# 1부터 10까지 더하기
n = 1
total = 0
while True:
    if n > 10:
        break
    total = total + n
    n = n + 1

print("합계는", total, "입니다")
```

```python
# for문

# 리스트 - 여러 개의 자료를 저장하는 자료구조이다.
cart = "라면"
print(cart)

carts = ["라면", "계란", "우유"]
print(carts)
print(type(carts))

# range(처음값, 종료값) 사용 
# 종료값은 종료값-1로 출력됨
print(range(10))
print(range(0, 10))

print(list(range(10)))
print(list(range(0, 10)))

# 1부터 10까지 출력하기
# 1씩 증가
for i in range(1, 11, 1):
    print(i)

# 2씩 증가
for i in range(1, 11, 2):
    print(i, end=' ')

# 1부터 10까지의 합계 구하기
total = 0
for i in range(1, 11):
    total = total + i

print("\n합계는", total, "입니다")
```

```python
# 리스트 - 여러 개의 자료를 저장하는 자료구조이다.
cart = "라면"
print(cart)

carts = ["라면", "계란", "우유"]
print(carts)
print(type(carts))

# 인덱싱
print(carts[0])
print(carts[2])
print(carts[-1])

# 슬라이싱(범위)
print(carts[0:3])
print(carts[0:])
print(carts[:])

print(carts[0:-1])
print(carts[0:2])

# for ~ in 구문
print("라면" in carts)
print("커피" in carts)

for cart in carts:
    print(cart)

# 요소 수정
carts[2] = "생수"

# 요소 추가(맨 뒤)
carts.append("쌀")
print(carts)

# 요소 삭제(맨 뒤)
carts.pop()
print(carts)
```

```python
# 리스트의 연산
numbers = [1, 2, 3, 4]
print(numbers)

print(numbers[2])
print(numbers[-1])

# 전체 조회
for number in numbers:
    print(number)

# 요소 더하기
total = 0
for number in numbers:
    total = total + number

# print("합계:", total)
print(f'합계: {total}')

# 요소의 평균
# avg = total / 4
avg = total / len(numbers)
print(f'평균: {avg}')
```

```python
# 딕셔너리
dictbox = {"name": "쿠키", "level": 5}
print(dictbox)
print(type(dictbox))

# 키, 값
print(dictbox.keys())
print(dictbox.values())

# 요소 조회
print(dictbox["name"])
print(dictbox["level"])

# 요소 추가
dictbox["id"] = "cookie"
print(dictbox)

# 값 변경(업데이트)
dictbox["level"] = 6
print(dictbox["level"])

# 요소 삭제
del dictbox["name"]
print(dictbox)

# 딕셔너리 생성 - 요소의 값이 리스트인 경우
cook = ["분식", "한식", "중식"]
count = [3, 4, 1]

store = {
    "요리": cook,
    "점포수": count
}
print(store)
print(type(store))

# "한식"에 접근
print(store["요리"][1])
print(store["점포수"][1])
```
