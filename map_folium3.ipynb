```python
from folium import *
import pandas as pd

seoul_map = Map(location = [37.55, 126.99], zoom_start = 12)

# 딕셔너리로 구현
map_info = [
    {"location": [37.54322, 126.951576], "popup": "공덕역"},
    {"location": [37.497952, 127.027619], "popup": "강남역"},
    {"location": [37.571731, 127.011069], "popup": "동대문역"}
]

for map in map_info:
    Marker(
        location = map["location"],
        popup = map["popup"], icon=Icon(color='red', icon='arrow-down'),
    ).add_to(seoul_map)

seoul_map

# 위도, 경도(공덕역, 강남역, 동대문역)
# locations = [
#     [37.54322, 126.951576], 
#     [37.497952, 127.027619],
#     [37.571731, 127.011069]
# ]

# for location in locations:
#     Marker(
#         location = location,
#         icon=Icon(color='blue', icon='arrow-down'),
#     ).add_to(seoul_map)
```
