```python
import pandas as pd

# 생성1
df = pd.DataFrame({
    "name": ['키보드', '마우스', '모니터'],
    "price": [17000, 25000, 78000],
    "spec": ['유선', '무선', '22inch']
})

# DataFrame 클래스의 주요 메서드(함수)
# 데이터프레임 정보 : int64 - 정수형, object - 문자형, Non-null(결측치 아님)
df.info()

# 통계 요약 - 기본적으로 수치형 칼럼이 출력됨
df.describe(include='number')
df.describe(include='object')

# 결측치 보기
df.isnull()
# True인 것의 개수
df.isnull().sum()

# 기본 5개 출력됨
df.head(2)

df.tail(2)

# 칼럼 이름 변경
df = df.rename(columns={"name":"상품", "price": "가격", "spec": "스펙"})
df

# 칼럼 추가(파생 칼럼) - 할인가격
df['할인가격'] = df['가격'] * 0.1
df
df.info()

# csv 파일로 저장 - to_csv()
df.to_csv("product.csv", index=False)

# csv 파일 읽어오기
pd.read_csv("product.csv")
```
