```python
import pydataset as pyds
import pandas as pd

mpg = pyds.data('mpg')
mpg

# mpg 데이터셋을 csv 파일로 만들기
mpg.to_csv("./datas/mpg", index=False)

# csv 파일 읽기
mpg = pd.read_csv("./datas/mpg")
mpg
```

```python
mpg.head()

mpg.tail()

mpg.info()

mpg.describe()
```

```python
# 칼럼명 추가하기 - 통합 연비
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()

# 통합 연비 평균
# sum(mpg['total']) / len(mpg)
# mpg['total'].mean()
```

```python
print(mpg['total'].describe())

# DataFrame의 plot을 사용하여 빈도 히스토그램 그래프
mpg['total'].plot.hist()
```

```python
import pandas as pd
import numpy as np

mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg
# mpg.info()
```

```python
# 빈도표로 합격 판정 자동차 수 살펴보기

count_test = mpg['test'].value_counts()
print(count_test)

count_test.plot.bar()
```
